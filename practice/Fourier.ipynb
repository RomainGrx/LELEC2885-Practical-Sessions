Original image
---

```python

img = imageio.imread('../images/cameraman.png')
nd_img = np.array(img)
```

Fourier transform
---

```python
W, H = nd_img.shape
fft_img = np.fft.fft2(nd_img)
fft_img_shifted = np.fft.fftshift(fft_img)
fourier_img = np.abs(fft_img_shifted)
```

```python
fig, (axL, axR) = plt.subplots(nrows=1, ncols=2, figsize=(10,8))

axL.set_title('Original image')
axL.imshow(nd_img, cmap='gray')

axR.set_title('Fourier domain image')
axR.imshow(20*np.log(fourier_img), cmap='gray')

plt.show()
```
